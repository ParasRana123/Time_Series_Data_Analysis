```python
import yfinance as yf

data = yf.download('TSLA' , start='2025-01-01' , end='2026-01-10')
```

```python
df_tesla = data
```

```python
df_tesla['High'].plot(figsize=(12 , 4))
```

```python
# providing the xlimit and the ylimit
df_tesla['High'].plot(xlim=['2025-01-01' , '2025-04-01']  , figsize=(12 , 4))
```

```python
# providing the xlimit and the ylimit
df_tesla['High'].plot(xlim=['2025-01-01' , '2025-04-01'] , ylim=[0 , 500] , figsize=(12 , 4))
```

```python
# providing the xlimit and the ylimit and colouring
df_tesla['High'].plot(xlim=['2025-01-01' , '2025-04-01']  , figsize=(12 , 4) , ls='--' , c='green')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
index = df_tesla.loc['2025-01-01' : '2025-08-01'].index
share_open = df_tesla.loc['2025-01-01' : '2025-08-01']['Open']
```

```python
figure , axis = plt.subplots()
plt.tight_layout()
figure.autofmt_xdate()       # Used for preventing the overlapping in the X-axis
axis.plot(index , share_open)
```
