### Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A')
```

```python
# Will give data only at the end of that year and will tell us that at the end of the day which was the min of close , high , low ...etc
df_tesla.resample(rule='A').min()
```

```python
# Gives us the year end frequency
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
# Rule QS: Quaterly Start Frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
# Business End Frequency
df_tesla.resample(rule='BA').max()
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar')
```

```python
df_tesla['High'].rolling(10).mean()
```

```python
df_tesla['Open 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```
